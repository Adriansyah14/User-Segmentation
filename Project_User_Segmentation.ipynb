```python
# pip install pandas
# pip install numpy
# pip install datetime

import datetime as dt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Adriansyah14/User-Segmentation/refs/heads/main/Salinan_Online_Retail_Data.csv")
df
```

```python
df.info()
```

# Cleaning Data

```python
# tampilkan missing value
df.isnull().sum()
```

```python
# buat salinan dataset
df_clean = df.copy()
```

```python
# hapus/drop semua baris yang tidak memiliki product_name
df_clean = df_clean[~df_clean["product_name"].isnull()]
```

```python
# hapus/drop semua baris yang tidak memiliki customer_id
df_clean = df_clean[~df_clean["customer_id"].isna()]
```

```python
df_clean.isnull().sum()
```

```python
# membuat kolom date
df_clean["date"] = pd.to_datetime(df_clean["order_date"]).dt.date.astype("datetime64[ns]")
```

```python
# ubah product_name jadi huruf kecil
df_clean["product_name"] = df_clean["product_name"].str.lower()
```

```python
# hapus semua baris dengan prdoct_code atau product_name_test
df_clean = df_clean[(~df_clean['product_code'].str.lower().str.contains('test')) |
                    (~df_clean['product_name'].str.contains('test '))]
```

```python
# buat kolom baru yaitu order_status dengan nilai "cancelled" jika pada order_id diawali huru 'C' dan "delivered" jika order_id tidak terdapat awalan huruf 'C'
df_clean['order_status'] = np.where(df_clean['order_id'].str[:1]=='C', 'cancelled', 'delivered')
```

```python
# mengubah nilai quantity yg negatif ke positif karena nilai negatif menandakan order cancelled
df_clean['quantity'] = df_clean['quantity'].abs()
```

```python
# hapus baris pada price bernilai negatif
df_clean = df_clean[df_clean['price']>0]
```

```python
# membuat kolom amount, yaitu perkalian antara quantity dan price
df_clean['amount'] = df_clean['quantity'] * df_clean['price']
```

```python
# mengganti product_name dari product_code yang memiliki beberapa product_name dengan salah satu product_name-nya yang paling sering muncul
most_freq_product_name = df_clean.groupby(['product_code','product_name'], as_index=False).agg(order_cnt=('order_id','nunique')).sort_values(['product_code','order_cnt'], ascending=[True,False])
most_freq_product_name['rank'] = most_freq_product_name.groupby('product_code')['order_cnt'].rank(method='first', ascending=False)
most_freq_product_name = most_freq_product_name[most_freq_product_name['rank']==1].drop(columns=['order_cnt','rank'])
df_clean = df_clean.merge(most_freq_product_name.rename(columns={'product_name':'most_freq_product_name'}), how='left', on='product_code')
df_clean['product_name'] = df_clean['most_freq_product_name']
df_clean = df_clean.drop(columns='most_freq_product_name')
```

```python
# mengkonversi tipe data customer_id menjadi string
df_clean['customer_id'] = df_clean['customer_id'].astype(str)
```

```python
# menghapus outlier
from scipy import stats
df_clean = df_clean[(np.abs(stats.zscore(df_clean[['quantity','amount']]))<3).all(axis=1)]
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
df_clean.info()
```

# Membuat RFM segmentation

## Agregat data transaksi ke bentuk summary total transaksi (order), total nilai order (order value), tanggal order terakhir dari setiap pengguna

```python
df_user = df_clean.groupby(['customer_id'], as_index=False).agg(order_cnt=('order_id','nunique'), total_order_value=('amount','sum'), max_order_date=('date','max'))
df_user
```

## Buat kolom jumlah hari terakhir order

```python
today = df_clean['date'].max()
```

```python
df_user['days_since_last_order'] = (today - df_user['max_order_date']).dt.days
df_user
```

```python
df_user.describe()
```

## Buat binning dari jumlah hari sejak order terakhir, yang terdiri dari 5 bins dengan batas min, P20, P40, P60, P80, max dan beri label 1-5 dari nilai bin tertinggi ke terendah ssebagai skor recency

```python
df_user['recency_score'] = pd.cut(df_user['days_since_last_order'],
                                  bins=[df_user['days_since_last_order'].min(),
                                        np.percentile(df_user['days_since_last_order'], 20),
                                        np.percentile(df_user['days_since_last_order'], 40),
                                        np.percentile(df_user['days_since_last_order'], 60),
                                        np.percentile(df_user['days_since_last_order'], 80),
                                        df_user['days_since_last_order'].max()],
                                  labels=[5, 4, 3, 2, 1],
                                  include_lowest=True).astype(int)
df_user
```

```python
# cek df_user setelah penambahan kolom
df_user.info()
```

## Buat binning dari total transaksi (order) yang terdiri dari 5 bins dengan batas-batasnya merupakan min, P20, P40, P60, P80, max dan beri label 1-5 dari bin terendah ke tertinggi sebagai skor frequency

```python
df_user['frequency_score'] = pd.cut(df_user['order_cnt'],
                                    bins=[0,
                                          np.percentile(df_user['order_cnt'], 20),
                                          np.percentile(df_user['order_cnt'], 40),
                                          np.percentile(df_user['order_cnt'], 60),
                                          np.percentile(df_user['order_cnt'], 80),
                                          df_user['order_cnt'].max()],
                                    labels=[1, 2, 3, 4, 5],
                                    include_lowest=True).astype(int)
```

```python
df_user
```

```python
df_user.info()
```

## Buat binning dari total nilai order (order value) yang terdiri dari 5 bins dengan batas-batasnya merupakan min, P20, P40, P60, P80, max dan beri label 1 sampai 5 dari bin terendah ke tertinggi sebagai skor monetary

```python
df_user['monetary_score'] = pd.cut(df_user['total_order_value'],
                                   bins=[df_user['total_order_value'].min(),
                                         np.percentile(df_user['total_order_value'], 20),
                                         np.percentile(df_user['total_order_value'], 40),
                                         np.percentile(df_user['total_order_value'], 60),
                                         np.percentile(df_user['total_order_value'], 80),
                                         df_user['total_order_value'].max()],
                                   labels=[1, 2, 3, 4, 5],
                                   include_lowest=True).astype(int)
df_user
```

```python
df_user.info()
```

## Buat kolom baru dengan nama 'segment' berdasarkan 'recency_score' dan 'frequency_score'

```python
df_user['segment'] = np.select(
    [(df_user['recency_score']==5) & (df_user['frequency_score']>=4),
     (df_user['recency_score'].between(3, 4)) & (df_user['frequency_score']>=4),
     (df_user['recency_score']>=4) & (df_user['frequency_score'].between(2, 3)),
     (df_user['recency_score']<=2) & (df_user['frequency_score']==5),
     (df_user['recency_score']==3) & (df_user['frequency_score']==3),
     (df_user['recency_score']==5) & (df_user['frequency_score']==1),
     (df_user['recency_score']==4) & (df_user['frequency_score']==1),
     (df_user['recency_score']<=2) & (df_user['frequency_score'].between(3, 4)),
     (df_user['recency_score']==3) & (df_user['frequency_score']<=2),
     (df_user['recency_score']<=2) & (df_user['frequency_score']<=2)],
    ['01-Champion', '02-Loyal Customers', '03-Potential Loyalists', "04-Can't Lose Them", '05-Need Attention',
     '06-New Customers', '07-Promising', '08-At Risk', '09-About to Sleep', '10-Hibernating']
)
df_user
```

```python
df_user.info()
```

## Tampilkan summary RMF segmentation

```python
summary = pd.pivot_table(df_user, index='segment',
               values=['customer_id','days_since_last_order','order_cnt','total_order_value'],
               aggfunc={'customer_id': pd.Series.nunique,
                        'days_since_last_order': [np.mean, np.median],
                        'order_cnt': [np.mean, np.median],
                        'total_order_value': [np.mean, np.median]})
summary['pct_unique'] = (summary['customer_id'] / summary['customer_id'].sum() * 100).round(1)
summary
```
